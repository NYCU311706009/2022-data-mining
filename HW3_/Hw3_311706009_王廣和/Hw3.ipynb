# Data Preprocessing

## 導入資料

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('新竹_2021.xls')
```

```python
df.head()
```

```python
df = df.drop([0])
```

## 把字串清乾淨，str都會有很多空白接著

```python
df.columns = (pd.Series(df.columns).apply(lambda x: x.strip(' ') if isinstance(x,str) else str(x)))
```

```python
df.head()
```

```python
df['測項'] = df['測項'].apply(lambda x: x.strip())
```

```python
feature_name = df['測項'].unique()
feature_name
```

```python
df.head()
```

## 篩選日期並調整size

```python
df = df.set_index('日期')
```

```python
df = df.loc['2021-10-01':'2021-12-31']
df
```

```python
data = pd.DataFrame()
for i in feature_name:
    temp_df = df[df['測項']== i]
    #temp_df = temp_df.set_index('日期')
    temp_reshape = pd.DataFrame(np.matrix(temp_df.drop(columns = ['測站','測項'])).reshape(-1))
    data = pd.concat([data,temp_reshape],axis=0)
```

```python
data = data.set_index(feature_name)
data = data.apply(pd.to_numeric,errors='coerce')
#data = data.fillna(0)
```

```python
#data.to_csv('test.csv')
```

## 填空

```python
nan_indices = np.column_stack([np.where(data.isnull())[0], np.where(data.isnull())[1]])
nan_indices[0]
```

```python
data.iloc[nan_indices[0][0],nan_indices[0][1]]
```

```python
for i in nan_indices: 
    front = 0.0
    behind = 0.0
    #print(i[0],i[1])
    # 找前一個不為空值的
    for x in range(1,len(data.iloc[0])):
        if np.isnan(data.iloc[i[0],i[1]-x]) == False:
            front =  data.iloc[i[0],i[1]-x]
            break;
    # 找後一個不為空值的    
    for x in range(1,len(data[0])):
        if np.isnan(data.iloc[i[0],i[1]+x]) == False:
            behind =  data.iloc[i[0],i[1]+x]
            break;
    data.iloc[i[0],i[1]] = (front+behind)/2    
    #print('front',front,'behind',behind)
    #print(data.iloc[i[0],i[1]])
```
